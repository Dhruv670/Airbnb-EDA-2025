```python
# DATA CLEANING 1 : Covert "last review" to DATETIME
import pandas as pd
df['last review'] = pd.to_datetime(df['last review'], errors='coerce', format='mixed')
df.info()

# This Command Preserves the Date-Values that are of INVALID DATE FORMATS hence, avoiding any ValueError and converts them into a common Format. Hence, avoids DATA LOSS.
```

```python
# Display first 10 records of 'price' column
print(df['price'].head(10))

```

```python
# Remove dollar signs and commas,Convert to numeric (float or int),Handle missing (NaN) values :

df['price'] = df['price'].replace('[\$,]', '', regex=True).replace('', None)
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Prints the Final Output of Updated Clean Values (Unitless).
print(df['price'].head(10))
```

```python
# Checks for NULL VALUES AND Returns BOOELAN OUTPUT.  Each cell is True if the VALUE is missing (NaN) and,
# Each cell is False if the VALUE is present
df.isnull()
```

```python
# ADDS UP AND PRODUCES THE LIST OF ALL THE COLUMNS WITH THEIR RESPECTIVE NULL-VALUE COUNTS

df.isnull().sum()
```

```python
# Drop column 'license' with the IF condition to ensure it doesn't throw an ERROR  upon random factors : Spelling mistakes.
if 'license' in df.columns:
  df.drop('license', axis=1, inplace =True)


```

```python
#Checks if the column 'license' in DELETED
df.head()
```

```python
df.dropna(subset =['neighbourhood', 'neighbourhood group'])
```

```python
# Calculate the average price for each neighborhood

average_price_by_neighborhood = df.groupby('neighbourhood')['price'].mean().reset_index()
average_price_by_neighborhood
```

```python
range_price_by_neighborhood = df.groupby('neighbourhood')['price'].apply(lambda x: x.max() - x.min()).reset_index()
range_price_by_neighborhood

```

```python
# RETURN TOP 10 MOST EXPENSIVE/PRICED NEIGHBOURHOOD TO TAKE AN AIRBNB



TOP_Neighbourhoods = average_price_by_neighborhood.sort_values (by ='price', ascending = False).head(10)
TOP_Neighbourhoods
```

```python
# RETURN TOP 10 MOST EXPENSIVE/PRICED NEIGHBOURHOOD TO TAKE AN AIRBNB WITH THE 'OLD INDEX REMOVED' :

# reset_index() → It resets the row index of a DataFrame back to the default integers (0, 1, 2, ...).


# drop=True → prevents old index from becoming a new column



TOP_Neighbourhoods = average_price_by_neighborhood.sort_values (by ='price', ascending = False).head(10).reset_index(drop = True)
TOP_Neighbourhoods

```

```python
# CREATE A VISUALISATION TO REPRESNT THE TOP 10 NEIGHBOURHOODS BY PRICE.

import matplotlib.pyplot as plt

# Plotting Top 10 Neighbourhoods by Average Price
plt.figure(figsize=(12, 6))
plt.bar(TOP_Neighbourhoods['neighbourhood'], TOP_Neighbourhoods['price'], color = 'orange')

plt.title('Top 10 Neighbourhoods by Average Price', fontsize=14)
plt.xlabel('Neighbourhood', fontsize=12)
plt.ylabel('Average Price ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```
